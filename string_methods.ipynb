# in string slicing syntax is :

```python
k="Vinoth kumar M "
print(len(k))
```

```python
k="Vinoth kumar M "
print(k[0:])
```

```python
k="Vinoth kumar"
print(k[-5:-2])
```

```python
k="Vinoth"
language="Python"
print(f"my name is {k} and im a {language} developer")
```
