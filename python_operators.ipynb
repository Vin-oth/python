# ASSIGNMENT OPERATORS

```python
x=5;
print(x)

x+=3
print(x)

x-=3
print(x)

x*=3
print(x)

x/=3
print(x)

x%=3
print(x)

y=x+10
y//=3
print(x)

x**=3
print(x)

x=3
x &= 2
print(x)

x=12
x |= 5
print(x)

```

# COMPARISON OPERATORS

```python
x=5
y=5
v=x==y
print(v)
```

```python
x=5
y=34
v=x!=y
print(v)
```

```python
x=5
y=9
print(x>y)
print(x<y)

print(x>=y)
print(x<=y)
```

# LOGICAL OPERATORS

```python
x=5
y=3

print(x>y and x!=y)
print(x>y or x!=y)
print(not(x>y and x!=y))
```
